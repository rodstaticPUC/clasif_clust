# Clasificacion y Clustering

Generación de conjuntos de datos para testear tecnicas de clustering y clasificacion.
- numero de muestras
- mezcla (shuffle)
- ruido
- parametros de aleatoriedad



```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
X,y = make_moons(n_samples=100, noise=0.1, random_state=10)
print(X.shape)
```

# Generación de 5 graficos con aumento del ruido

Aumento del ruido cada 0.04

```python
fig, axs = plt.subplots(1,5,figsize=(18,3)) #configuracion grafica

for i, ax in zip(range(5),axs):
    X,y =make_moons(n_samples=100, noise=i*0.04,random_state=10)
    for target, color, marker in zip(range(2),['r','k'],['o','^']):
        ax.scatter(X[y==target,0],X[y==target,1],c=color,marker=marker,label='class'+format(target))
        ax.set_xlabel('caracteristica 1',fontsize=20)
        if (i==0):
            ax.set_ylabel('caracteristica 2',fontsize=20)
        ax.set_title('Moons, noise: '+format(i*0.04))
        ax.grid()
plt.show()
```
